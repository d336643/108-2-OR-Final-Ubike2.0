```python
raw = pd.read_csv("youbike.csv")
raw=raw.rename(columns={'Unnamed: 1':"站點"})
raw=raw.rename(columns={'Unnamed: 2':"車柱數量"})
savedCol = ['站點',"車柱數量"]
for col in raw.columns:
    if raw[col][0] == "變動量":
        savedCol.append(col)
df = raw[savedCol]
pickedStop = ['捷運公館站(2號出口)', '臺大總圖書館西南側', '臺大社科院西側']
data = {}
for stop in pickedStop:
    tmp = [int(x) for x in (df[df['站點'] == stop]).values[0][1:]]
    data[stop] = tmp
data
```

```python
import requests
import json
import pickle
import time
```

```python
cnt = 60
while(cnt<120):
    url = "https://apis.youbike.com.tw/api/front/station/all?lang=zh&type=2"
    data = requests.get(url).json()
    data = [_dict for _dict in data["retVal"] if _dict['area_code'] == "00"]
    current_time = data[0]['time']

    save_data = []
    for stop_info in data:
        save_data.append({'time':stop_info['time'], 'name_tw':stop_info['name_tw'],'parking_spaces':stop_info['parking_spaces'],
                                       'available_spaces':stop_info['available_spaces'],'empty_spaces':stop_info['empty_spaces'],
                                       'updated_at':stop_info['updated_at']})

    f = open("20200617data/"+str(cnt)+".pkl","wb")
    pickle.dump(save_data,f)
    f.close()
    time.sleep(300)
    cnt+=1
```
